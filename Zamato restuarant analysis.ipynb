```python
# Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.rcParams['figure.figsize']=(10,6)
sns.set(style="darkgrid")
```

```python
# Main csv file
zamato = pd.read_csv('zomato.csv', encoding='latin-1')
```

```python
# Secondary file
country_code = pd.read_excel('Country-Code.xlsx')
```

```python
# Merge the DataFrames
df = pd.merge(zamato, country_code, on='Country Code', how='left')
```

```python
country_count =df['Country'].value_counts().reset_index()
country_count.rename(columns={'index':'Country', 'Country':'Count'}, inplace = True)
country_count
```

* The country with the most restaurants is India with 8652 restaurants, followed by the United States with 434 restaurants. Canada has the fewest restaurants in the dataset with only 4 establishments.

```python
# A barplot
sns.barplot(data=country_count, x='Country', y='Count')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Restaurant Distribution: Top Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Value count of city restuarant
city_count = df['City'].value_counts().head(10).reset_index()
city_count.rename(columns={'index':'City', 'City':'Count'}, inplace = True)
city_count
```

* New Delhi has the highest number of restaurants, with 5473 establishments, followed by Gurgaon and Noida with 1118 and 1080 restaurants, respectively. The count gradually decreases for the subsequent cities. 

```python
# A barplot
sns.barplot(data=city_count, x='City', y='Count')
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Restaurant Distribution: Top Cities')
plt.xticks(rotation=45)
plt.show()
```

```python
city_cuisines = df['Cuisines'].value_counts().head(10).reset_index()
city_cuisines.rename(columns={'index':'Cuisines', 'Cuisines':'Count'}, inplace = True)
city_cuisines
```

* "North Indian" cuisine appears to be the most prevalent with 936 restaurants serving it. The combination of "North Indian, Chinese" is the second most common, followed by "Chinese" and "Fast Food" cuisines, each with 354 restaurants. While Street Food 149 is least amount the top 10 cuisine

```python
# A barplot
sns.barplot(data=city_cuisines, x='Cuisines', y='Count')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.title('Restaurant Distribution: Top Cuisines')
plt.xticks(rotation=45)
plt.show()
```

```python
ratings_count = df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
ratings_count.rename(columns={0: 'Rating count'}, inplace=True)
ratings_count
```

```python
# Plot barplot
cols = ['white','red','orange','yellow','green','darkgreen']
sns.barplot(data=ratings_count, x="Aggregate rating", y="Rating count")
plt.xlabel("Aggregate rating")
plt.ylabel("Count")
plt.title("Restaurant Rating Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot barplot
cols = ['white','red','orange','yellow','green','darkgreen']
sns.barplot(data=ratings_count, x="Aggregate rating", y="Rating count", hue='Rating text', palette=cols)
plt.xlabel("Aggregate rating")
plt.ylabel("Count")
plt.title("Restaurant Rating Distribution")
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

```python
# Plot barplot
sns.barplot(data=ratings_count, x="Aggregate rating", y="Rating count",hue='Rating color', palette=cols)
plt.xlabel("Aggregate rating")
plt.ylabel("Count")
plt.title("Restaurant Rating Distribution")
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.show()
```
